## **Group By**

#### The groupby() function in pnadas is used to group data based on one or more columns in a DataFrame

```python
df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
#now constructing the graph 
Sqty=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=Sqty)
```

##### *Note: From above Graph you can see that most of the Qty buys M-Size in the sales*

```python
#as the CAtegory datatype is object so we need to change it to String
df['Category']=df['Category'].astype(str)
column_data=df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=35,edgecolor='Cyan')
plt.xticks(rotation=90)
plt.title('Category wise Bar PLot Data')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

##### *Note: From the above Graph we can see that most of the buyers are T-shirt*

```python
#Checking Size Data by Pie Chart
Size_Data=df['Size'].value_counts()
plt.pie(Size_Data,labels=Size_Data.index,autopct='%1.1f%%',pctdistance=0.85,labeldistance=1.1)
plt.axis('equal')
plt.show()
```

```python
#Checking Size Data by Pie Chart
Size_Data=df['Size'].value_counts()
plt.pie(Size_Data,labels=Size_Data,autopct='%1.1f%%',pctdistance=0.85,labeldistance=1.1)
plt.axis('equal')
plt.show()
```

```python
#Prepare data for scatter plot
x_data=df['Category']
y_data=df['Size']

#Plot the scatter plot
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show();
```

```python
#Prepare data for pie chart
a1=df['Fulfilment'].value_counts();

fig, ax=plt.subplots()

ax.pie(a1,labels=a1.index,autopct='%1.1f%%',radius=0.7,wedgeprops=dict(width=0.6))
ax.set(aspect="equal")

plt.show()
```

```python
#Plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Ditribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 10 states
top_10_states=df['ship-state'].value_counts().head(10)
#Plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_states.index)],x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=45)
plt.show()
```

##### *Note: From above Graph we can see tha most of the buyers are from the Maharashtra state*

```python
sales_by_category = df.groupby('Category')['Amount'].sum().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='Category', y='Amount', data=sales_by_category, palette='viridis')
plt.title('Total Sales by Category')
plt.xticks(rotation=45)
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Scatter plot of Sales vs Quantity
plt.figure(figsize=(5,6))
sns.scatterplot(x='Qty', y='Amount', data=df, hue='Category', palette='bright', alpha=0.6)
plt.title('Sales vs. Quantity')
plt.xlabel('Quantity')
plt.ylabel('Sales Amount')
plt.legend(loc='upper right')
plt.show()
```
